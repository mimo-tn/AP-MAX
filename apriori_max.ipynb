```python
def generate_frequent_1_itemsets(dataset, min_support):
    frequent_1_itemsets = {}
    for transaction in dataset:
        for item in transaction:
            if item in frequent_1_itemsets:
                frequent_1_itemsets[item] += 1
            else:
                frequent_1_itemsets[item] = 1

    frequent_1_itemsets = {item: support for item, support in frequent_1_itemsets.items() if support >= min_support}
    return frequent_1_itemsets
def generer_combinaisons(liste):
    combinaisons = []
    if(len(liste)>0):
        for r in range(len(liste) - 1, len(liste)):
            combinaisons.extend(list(itertools.combinations(liste, r)))
    combinaisons_uniques = set(combinaisons)
    return combinaisons_uniques
def generer_combinaisons_recur(liste):
    combinaisons = []
    for r in range(len(liste) - 1, len(liste)):
        combinaisons.extend(list(itertools.combinations(liste, r)))
    combinaisons_uniques = set(combinaisons)
    return combinaisons_uniques
def generate_candidate_itemsets(prev_frequent_itemsets, k):
    candidate_itemsets = []
    for i in range(len(prev_frequent_itemsets)):
        for j in range(i + 1, len(prev_frequent_itemsets)):
            itemset1 = prev_frequent_itemsets[i]
            itemset2 = prev_frequent_itemsets[j]
            if itemset1[:-1] == itemset2[:-1]:
                new_itemset = itemset1.join(itemset2[-1])
                candidate_itemsets.append(new_itemset)
    return candidate_itemsets

#def prune_infrequent_itemsets(candidate_itemsets, frequent_itemsets):
#    pruned_itemsets = []
#    for itemset in candidate_itemsets:
#        is_infrequent = False
#        for i in range(len(itemset)):
#            subset = itemset[:i].join(itemset[i + 1:])
#            if subset not in frequent_itemsets:
#                is_infrequent = True
#                break
#        if not is_infrequent:
#            pruned_itemsets.append(itemset)
#    return pruned_itemsets

def calculate_support(dataset, itemsets):
    itemset_counts = {}
    for transaction in dataset:
        for itemset in itemsets:
            if set(itemset).issubset(transaction):
                if tuple(itemset) in itemset_counts:
                    itemset_counts[tuple(itemset)] += 1
                else:
                    itemset_counts[tuple(itemset)] = 1
    return {itemset: support for itemset, support in itemset_counts.items()}

def calculate_support_rec(dataset, itemset):
    count = 0
    for transaction in dataset:
        if itemset.issubset(transaction):
            count += 1
    return count


```

```python
def find_frequent_itemsets(dataset, itemsets, min_sup, current_itemset=None, frequent_itemsets=None):
    if frequent_itemsets is None:
        frequent_itemsets = {}

    if current_itemset is None:
        current_itemset = set()

    # Count support for the current itemset
    support = calculate_support_rec(dataset, current_itemset)
    if support >= min_sup:
        frequent_itemsets[tuple(current_itemset)] = support

    # Generate candidate itemsets by extending the current itemset
    candidate_itemsets = generer_combinaisons_recur(list(current_itemset))

    for itemset in candidate_itemsets:
        # Recursively find frequent itemsets
        updated_itemset = current_itemset.copy()
        updated_itemset.add(itemset)
        frequent_itemsets = find_frequent_itemsets(dataset, itemsets, min_sup, updated_itemset, frequent_itemsets)

    return frequent_itemsets
```
